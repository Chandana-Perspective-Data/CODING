```python
import pandas as pd
import matplotlib.pyplot as plt

# create the dataframe
data = {
    'Name': ['spoorthi', 'pallavi', 'anisha feebi', 'Garvitha', 'mary akhila',
             'Deepthi', 'harshitha', 'srujana', 'venus', 'vaishnavi',
             'Bargavi', 'navya', 'harshitha', 'pravalika', 'nandhini'],
    'Age': [15, 15, 12, 10, 18, 20, 20, 20, 19, 19, 15, 15, 14, 13, 13],
    'offline/online': ['offline', 'offline', 'online', 'offline', 'offline',
                       'offline', 'offline', 'offline', 'online', 'offline',
                       'online', 'online', 'offline', 'offline', 'online'],
    'mobile used': ['parents', 'own', 'parents', 'parents', 'parents',
                    'parents', 'own', 'parents', 'parents', 'parents',
                    'parents', 'parents', 'own', 'own', 'own'],
    'gender': ['F', 'F', 'F', 'F', 'F', 'F', 'F', 'F', 'F', 'F', 'F', 'F',
               'F', 'F', 'F'],
    'study': ['Bipc', 'MEC', 'MPC', 'Bipc', 'MPC', 'CEC', 'Bipc', 'MEC',
              'Bipc', 'MPC', 'Bipc', 'MEC', 'MEC', 'MEC', 'MPC']
}
df = pd.DataFrame(data)
df
```

```python
# 1. Grouping the data by the 'study' column and finding the average age of each group.
avg_age_by_study = df.groupby('study')['Age'].mean()
print(avg_age_by_study)
```

```python
# 2. Creating a new column called 'phone owner' that indicates whether the student owns their mobile device or not.
df['phone owner'] = df['mobile used'].apply(lambda x: 'own' if x == 'own' else 'not own')
print(df)
```

```python
# 3. Filtering the data to only include students who are studying 'MPC' and are using their own mobile device.
mpc_own = df[(df['study'] == 'MPC') & (df['mobile used'] == 'own')]
print(mpc_own)
```

```python
# 4. Counting the number of students in each gender and offline/online category.
count_by_group = df.groupby(['gender', 'offline/online']).size()
print(count_by_group)
```

```python
# 5. Sorting the data by age in ascending or descending order.
df_sorted = df.sort_values('Age', ascending=True)
print(df_sorted)
```

```python
# 6. Finding the maximum and minimum age values in the dataset.
max_age = df['Age'].max()
min_age = df['Age'].min()
print(f"Max age: {max_age}, Min age: {min_age}")
```

```python
# 7. Creating a bar chart to visualize the number of students in each study program.
study_count = df['study'].value_counts()
study_count.plot(kind='bar')
plt.show()
```

```python
# Scatterplot showing the relationship between age and study program:
import matplotlib.pyplot as plt

# scatterplot
plt.scatter(df['Age'], df['study'])

# add labels and title
plt.xlabel('Age')
plt.ylabel('Study program')
plt.title('Age vs Study program scatterplot')

# show the plot
plt.show()
```

```python
# Pie chart showing the distribution of offline/online usage:
# count the number of offline/online users
offline_count = df[df['offline/online'] == 'offline'].shape[0]
online_count = df[df['offline/online'] == 'online'].shape[0]

# create a pie chart
plt.pie([offline_count, online_count], labels=['Offline', 'Online'], autopct='%1.1f%%')

# add title
plt.title('Offline/Online usage')

# show the plot
plt.show()
```

```python
# Line chart showing the average age of students in each study program:
# group data by study program and calculate the mean age
study_age = df.groupby('study')['Age'].mean()

# create a line chart
plt.plot(study_age.index, study_age.values)

# add labels and title
plt.xlabel('Study program')
plt.ylabel('Average age')
plt.title('Average age by study program')

# show the plot
plt.show()
```

```python
# Histogram showing the distribution of age:
# create a histogram
plt.hist(df['Age'], bins=10)

# add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age distribution')

# show the plot
plt.show()
```

```python
# Bar chart showing the number of students in each study program:
# count the number of students in each study program
study_counts = df['study'].value_counts()

# create a bar chart
plt.bar(study_counts.index, study_counts.values)

# add labels and title
plt.xlabel('Study program')
plt.ylabel('Number of students')
plt.title('Number of students by study program')

# show the plot
plt.show()
```

```python
# Scatter plot of age vs. study program:
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
colors = {'Bipc': 'r', 'MEC': 'g', 'MPC': 'b', 'CEC': 'y'}

for study, group in df.groupby('study'):
    ax.scatter(group['Age'], [study]*len(group), c=colors[study], label=study)

ax.legend()
ax.set_xlabel('Age')
ax.set_ylabel('Study Program')
plt.show()
```

```python
# Pie chart of number of students by gender
import matplotlib.pyplot as plt

gender_counts = df['gender'].value_counts()
labels = gender_counts.index
sizes = gender_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Gender Distribution')
plt.show()
```

```python
# Line plot of age vs. mobile ownership:
import matplotlib.pyplot as plt

mobile_counts = df['mobile used'].value_counts()
own = df[df['mobile used'] == 'own']
not_own = df[df['mobile used'] != 'own']
own_age_counts = own['Age'].value_counts().sort_index()
not_own_age_counts = not_own['Age'].value_counts().sort_index()

plt.plot(own_age_counts.index, own_age_counts.values, label='Own')
plt.plot(not_own_age_counts.index, not_own_age_counts.values, label='Not Own')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.title('Age vs. Mobile Ownership')
plt.show()
```

```python
# Histogram of age distribution:
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=range(10, 22), edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.title('Age Distribution')
plt.show()
```

```python
# Bar chart of number of students in each study program:

import matplotlib.pyplot as plt

study_counts = df['study'].value_counts()
labels = study_counts.index
sizes = study_counts.values

plt.bar(labels, sizes)
plt.xlabel('Study Program')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each Study Program')
plt.show()
```
